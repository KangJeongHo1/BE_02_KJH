```python
# Classification - 분류 알고리즘을 사용해서 꽃의 종류를 알아보자
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
iris.data   # feature data
iris.target # target data
```

```python
# 1. 준비되어있는 데이터 셋을 나눈다.
# (1) 학습용 데이터(train_data) (2) 테스트용 데이터(test_data)
# train_test_split > 학습용 데이터셋과 테스트용 데이터 셋을 나눈다
from sklearn.model_selection import train_test_split

# X_train: 학습용 피처 데이터, X_test: 테스트용 피처 데이터
# y_train: 학습용 레이블 데이터, y_test: 테스트용 레이블 데이터

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=101)
```

```python
len(y_test)  # 30개
len(y_train)    # 120개
```

```python
# 2. 학습을 위한 알고리즘을 불러온다
from sklearn.tree import DecisionTreeClassifier # 분류 알고리즘(supervised learning

dt_clf = DecisionTreeClassifier(random_state=101)
dt_clf
```

```python
# 3. 학습을 진행 > train_data (X_train, y_train) >> 모의고사
dt_clf.fit(X_train, y_train)    # fit() - 학습을 진행
```

```python
# 4. 학습을 실행 > text_data(X_test, y_test) >> 수능
# predict() > 예측
predict = dt_clf.predict(X_test)
predict
```

```python
# 5. 채점을 진행
# accuracy_score()

from sklearn.metrics import accuracy_score

accuracy_score(y_test, predict)
```

```python
df = pd.DataFrame(data=X_test, columns=iris.feature_names)

df['answer'] = y_test
df['predict'] = predict

df
```
