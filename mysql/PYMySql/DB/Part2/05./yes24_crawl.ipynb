```python
# webdriver-manager 라이브러리 불러오기
from webdriver_manager.chrome import ChromeDriverManager
ChromeDriverManager().install()
```

```python
from selenium import webdriver
browswer = webdriver.Chrome()

```

```python
url = 'https://www.google.com'
browswer.get(url)
```

```python
from selenium.webdriver.common.by import By
browswer.find_element(By.CLASS_NAME, 'uU7dJb').text

# <div class="uU7dJb">대한민국</div>
```

```python
url = 'https://www.yes24.com/Product/category/bestseller?CategoryNumber=001&sumgb=06'
browswer.get(url)
```

```python
# 1 페이지의 한개의 링크 데이터 전부 수집
# browswer.find_element(By.CLASS_NAME, "gd_name").get_attribute('href') << 요소 형태로 들어옴

# 1페이지 전체의 링크 데이터 수집
datas = browswer.find_elements(By.CLASS_NAME, "gd_name") # << list 형태로 들어옴

for i in datas:
    print(i.get_attribute('href'))  # << list형태를 요소 형태로 바꿔줌
```

```python
# 3 페이지 까지의 링크 데이터 전부 수집
# https://www.yes24.com/Product/Category/BestSeller?categoryNumber=001&pageNumber=1&pageSize=24
# https://www.yes24.com/Product/Category/BestSeller?categoryNumber=001&pageNumber=2&pageSize=24
# https://www.yes24.com/Product/Category/BestSeller?categoryNumber=001&pageNumber=3&pageSize=24
import time

link_list = []
for i in range(1, 4):
    print("*" * 10, f"현재 {i}페이지를 수집 중입니다.")
    url = f"https://www.yes24.com/Product/Category/BestSeller?categoryNumber=001&pageNumber={i}&pageSize=24"
    browswer.get(url)


    datas = browswer.find_elements(By.CLASS_NAME, "gd_name") # << list 형태로 들어옴

    for i in datas:
        link = print(i.get_attribute('href')) 
        link_list.append(link)

    time.sleep(3)

print(link_list)
```
