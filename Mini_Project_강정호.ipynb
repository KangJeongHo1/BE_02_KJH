```python
words = ["apple", "banana", "orange", "grape", "lemon"]

def display_word(word, guess_letter):
    display = ""                        #ìµœì¢…ì ìœ¼ë¡œ êµ¬ì„±í•  ë‹¨ì–´ë¥¼ ì§€ì¹­í•˜ëŠ” ë³€ìˆ˜
    for letter in word:                 #ì…ë ¥ëœ wordì— ê° ê¸€ìë¥¼ í™•ì¸í•©ë‹ˆë‹¤.
        if letter in guess_letter:      #ë§Œì•½ ì…ë ¥ëœ ê¸€ìê°€ guess_letterì— ìˆë‹¤ë©´
            display += letter + " "     #dlsplayì— í˜„ì¬ ê¸€ìë¥¼ ì¶”ê°€í•˜ê³  ë‹¨ì–´ ì‚¬ì´ì— ê°„ê²©ì„ ìœ„í•´ ê³µë°±ì„ ì¶”ê°€.
        else:
            display += "_ "             #í˜„ì¬ ê¸€ìê°€ ì‹œë„í•œ ê¸€ìì— ì—†ìœ¼ë©´ ë°‘ì¤„ì„ ì¶”ê°€

    return display.strip()              #ìµœì¢…ì ìœ¼ë¡œ displayë¥¼ ë°˜í™˜í•˜ê³  ì•ë’¤ ê³µë°± ì œê±° = stripì€ ê³µë°±ì œê±°

def hangman_game(word):
    max = 5
    guess_letter = []                   #ì‚¬ìš©ìê°€ ì‹œë„í•œ ê¸€ìë¥¼ ì¶”ì í•˜ëŠ” ë¦¬ìŠ¤íŠ¸
    max_left = max                      #ë‚¨ì€ ì‹œë„ë¥¼ ë‚˜íƒ€ë‚´ê¸° ìœ„í•œ ë³€ìˆ˜

    print("ê²Œì„ì„ ì‹œì‘í•©ë‹ˆë‹¤!!")

    while max_left > 0:
        guess = input("í•œ ê¸€ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: ").lower()  #lower < ëŒ€ë¬¸ìë¡œ ì…ë ¥í–ˆì„ ê²½ìš° ì†Œë¬¸ìë¡œ ë°”ê¾¸ê¸°

        #í•œ ê¸€ìë§Œ ì…ë ¥í•˜ì§€ ì•Šì€ ê²½ìš°
        if len(guess) != 1 or not guess.isalpha():  #guess.isalphaëŠ” ì•ŒíŒŒë²³ì„ ì°¾ê¸° ìœ„í•œ ë©”ì„œë“œ
            print("í•œ ê¸€ìë§Œ ì…ë ¥í•˜ì„¸ìš”.")
            continue

        #ì´ë¯¸ ì‹œë„í•œ ê¸€ìì¸ì§€ í™•ì¸í•˜ê¸°
        if guess in guess_letter:
            print("ì´ë¯¸ ì‹œë„í•œ ê¸€ìì…ë‹ˆë‹¤.")
            continue

        guess_letter.append(guess)                  #ì‹œë„í•œ ê¸€ìë¥¼ ì¶”ê°€í•œë‹¤.

        #ì •ë‹µì´ í¬í•¨ëœ ê²½ìš°
        if guess in word:
            print("ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨êµ°ìš”. ë” í˜ë‚´ë´ìš”!")
        else:
            max_left -=1
            print(f"í‹€ë ¸ìŠµë‹ˆë‹¤. ë‚¨ì€ ì‹œë„ íšŸìˆ˜: {max_left}")

        current_dis = display_word(word, guess_letter)  #í˜„ì¬ê¹Œì§€ ë§ì¶˜ ì•ŒíŒŒë²³ ê°€ì ¸ì˜¤ê¸°
        print(current_dis)

        #ëª¨ë“  ê¸€ìë¥¼ ë§ì¶˜ ê²½ìš°
        if "_" not in current_dis:
            print("ì¶•í•˜ë“œë¦½ë‹ˆë‹¤ ì •ë‹µì…ë‹ˆë‹¤!!!!")
            break

    #ì‹œë„ íšŸìˆ˜ë¥¼ ëª¨ë‘ ì‚¬ìš©í•œ ê²½ìš°
    if max_left == 0:
        print(f"ê²Œì„ ë! ì •ë‹µì€ '{word}'ì…ë‹ˆë‹¤.")

#ëœë¤ ë‹¨ì–´ ì„ íƒ
import random
selected_word = random.choice(words)

#ê²Œì„ ì‹œì‘
hangman_game(selected_word)
```
